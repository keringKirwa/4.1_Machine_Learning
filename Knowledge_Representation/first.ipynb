```python
import networkx as nx 

my_graph = nx.Graph()

my_graph.add_node(3, label="Node 3", color="blue")
my_graph.add_node(4, label="Node 4", color="red")


my_graph.add_edge(3,4)

print(my_graph.nodes())

nx.draw(
    G=my_graph,
    with_labels=True,
    node_color= "blue",
    font_weight= "bold",   
)

print(list(my_graph.nodes))
```
