```python
import matplotlib.pyplot as plt
```

```python

x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 1, 3, 7]

fig ,(ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize = (14,7))

ax1.plot(x_data,y_data,c="yellow", linestyle= "--")
ax1.set_xlabel('Time(t)')
ax1.set_ylabel('Price(Ksh)')
ax1.set_title('Simple Line Plot')

ax1.legend(["Student performance over time ", "Price in Ksh "])

ax2.plot(y_data,x_data,c="blue")
plt.show()
```
